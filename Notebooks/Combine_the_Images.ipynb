```
from glob import glob
import os
from tqdm import notebook
```

```
train_paths = [
 '/content/800_dataset_with_semantic_segmentation/train_images/*',
 '/content/1000_dataset_with_semantic_segmentation/train_images/*', 
 '/content/600_dataset_with_semantic_segmentation/train_images/*', 
 '/content/dataset_with_semantic_segmentation/train_images/*', 
]
segment_paths = [
 '/content/800_dataset_with_semantic_segmentation/train_segmentation/*',
 '/content/1000_dataset_with_semantic_segmentation/train_segmentation/*', 
 '/content/600_dataset_with_semantic_segmentation/train_segmentation/*', 
 '/content/dataset_with_semantic_segmentation/train_segmentation/*', 
]
```

```
import os
# Function to rename multiple files
def main():
   i = 0
   path="/content/dataset_with_semantic_segmentation/train_segmentation/"
  #  dest = "/content/dataset/train_images/"
   for filename in notebook.tqdm(os.listdir(path)):
      my_dest = "normal_segment"+str(i) + ".png"
      my_source = path + filename
      my_dest = path + my_dest
      # rename() function will
      # rename all the files
      os.rename(my_source, my_dest)
      i += 1
# Driver Code
if __name__ == '__main__':
   # Calling main() function
   main()
```

```
import shutil
for path in train_paths:
  for img in notebook.tqdm(glob(path)):
      shutil.copy(img, '/content/dataset/train_images/')
print(f"Train Images : {len(os.listdir('/content/dataset/train_images/'))}")
for path in segment_paths:
  for img in notebook.tqdm(glob(path)):
      shutil.copy(img, '/content/dataset/train_segmentation/')
print(f"Segmentation Images : {len(os.listdir('/content/dataset/train_segmentation/'))}")
```
